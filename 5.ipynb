```python
def upsample(features, target, repeat):
    features_zeros = features[target == 0]
    features_ones = features[target == 1]
    target_zeros = target[target == 0]
    target_ones = target[target == 1]

    features_upsampled = pd.concat([features_zeros] + [features_ones] * repeat)
    target_upsampled = pd.concat([target_zeros] + [target_ones] * repeat)
    
    features_upsampled, target_upsampled = shuffle(
        features_upsampled, target_upsampled, random_state=12345)
    
    return features_upsampled, target_upsampled
```

```python
def downsample(features, target, fraction):
    features_zeros = features[target == 0]
    features_ones = features[target == 1]
    target_zeros = target[target == 0]
    target_ones = target[target == 1]

    features_downsampled = pd.concat(
        [features_zeros.sample(frac=fraction, random_state=12345)] + [features_ones])
    target_downsampled = pd.concat(
        [target_zeros.sample(frac=fraction, random_state=12345)] + [target_ones])
    
    features_downsampled, target_downsampled = shuffle(
        features_downsampled, target_downsampled, random_state=12345)
    
    return features_downsampled, target_downsampled
```

```python
features_downsampled, target_downsampled = downsample(features_train, target_train, diff1)

model = (RandomForestClassifier(max_depth=12,
                                n_estimators=28,
                                random_state=12345).fit(features_downsampled,
                                                        target_downsampled)
        )

#model = (RandomForestClassifier(max_depth=12, n_estimators=28, random_state=12345, class_weight='balanced').fit(features_upsampled, target_upsampled))
#print(f1_score(target_valid, best_model.predict(features_valid)))
#model = LogisticRegression(random_state=12345, solver='liblinear').fit(features_downsampled, target_downsampled)

print(f1_score(target_valid, model.predict(features_valid)))
```

```python
#diff & diff1 - –æ—Ç–Ω–æ—à–µ–Ω–∏—è —Ä–∞–∑–º–µ—Ä–æ–≤ —Ç–∞–±–ª–∏—Ü —Å —Ü–µ–ª–µ–≤—ã–º –ø—Ä–∏–∑–Ω–∞–∫–æ–º 1 –∏ 0
diff_matrix = []
diff1_matrix = []
for i in range(1,4):
    diff_matrix.append(i)
    k = diff1 * i
    diff1_matrix.append(k)
print(diff_matrix, diff1_matrix)
```

```python
result = 0
for i,j in zip(diff_matrix, diff1_matrix):
    features_upsampled, target_upsampled = upsample(features_train, target_train, i)
    features_downsampled, target_downsampled = downsample(features_upsampled, target_upsampled, j)
    model = (RandomForestClassifier(max_depth=12,
                                    n_estimators=28,
                                    random_state=12345).fit(features_downsampled,
                                                                 target_downsampled)
            )
    if f1_score(target_valid, model.predict(features_valid)) > result:
        result = f1_score(target_valid, model.predict(features_valid))
        best_model = model
        ii=i
        jj=j
print(best_model, result,ii,jj)
```
