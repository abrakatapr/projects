### Расчет обогощения (MAE)

````python
#C — доля золота в концентрате после флотации/очистки;
#F — доля золота в сырье/концентрате до флотации/очистки;
#T — доля золота в отвальных хвостах после флотации/очистки.
def recovery(c, f, t):
    return ((c * (f-t)) / (f * (c-t))) * 100
````

````python
rougher_recovery_train = recovery(data_train['rougher.output.concentrate_au'],
                         data_train['rougher.input.feed_au'],
                         data_train['rougher.output.tail_au'])
````

````python
mean_absolute_error(data_train['rougher.output.recovery'], rougher_recovery_train)
````

## Модель

### Функция для вычесления sMAPE

````python
def smape(x, y):
    return 100/len(x) * np.sum(2 * np.abs(x - y) / (np.abs(x) + np.abs(y)))
````

````python
def final_smape(a,b):
    score = 0.25 * a + 0.75 * b
    return score
````

### Обработка данных для моделей

````python
data_train = data_train.interpolate(method='linear')
````

````python
data_test = data_test.interpolate(method='linear')
````

#### Модель для флотации

````python
#признаки принадлежащие первому этапу
data_train_rough = data_train.drop(data_train.columns[24:], axis=1).drop('date', axis=1)
data_test_rough = data_test.drop(data_train.columns[24:], axis=1).drop('date', axis=1)
````

````python
features_train = data_train_rough.drop(['rougher.output.recovery'], axis=1)
target_train = data_train_rough['rougher.output.recovery']
features_test = data_test_rough.drop(['rougher.output.recovery'], axis=1)
target_test = data_test_rough['rougher.output.recovery']
````

````python
smape_score = make_scorer(smape, greater_is_better=False)
````

````python
num_features = features_train.select_dtypes(include=np.number).columns.tolist()
num_steps = [('scaler', StandardScaler())]
num_transformer = Pipeline(num_steps)
preprocessor = ColumnTransformer(
    transformers=[
        ('num', num_transformer, num_features)
    ]
)
steps = [('preprocessor', preprocessor), ('regressor', RandomForestRegressor(random_state=12345))]
````

````python
params = [{'regressor': [LinearRegression()]},
          {'regressor': [Ridge(random_state=12345)],
           'regressor__alpha': [1, 5, 10]
          },
          {'regressor': [Lasso(random_state=12345)],
           'regressor__alpha': [1, 5, 10]
          },
          {'regressor': [RandomForestRegressor(random_state=12345)],
           'regressor__max_depth': range(4,8),
           'regressor__n_estimators': range(4,8)
          }
         ]
````

````python
def pipeline_gridsearchcv(X, y, steps, params, scoring, verbose = 3):
    pipe = Pipeline(steps)
    cv = KFold(n_splits=5, shuffle=False)
    grid = GridSearchCV(pipe, param_grid=params, cv=cv, scoring=scoring, verbose=verbose, error_score='raise')
    grid.fit(X, y)
    print('')
    print('Параметры лучшей модели:', grid.best_params_)
    print('Значение лучшей метрики качества:', -grid.best_score_)
    return grid
````

````python
grid_base = pipeline_gridsearchcv(features_train, target_train, steps, params, smape_score, verbose = 1)
````

````python
smape(target_train, grid_base.predict(features_train))
````

````python
smape(target_train, pd.Series(target_train.median(), index=target_train.index))
````

````python
smape(target_test, grid_base.predict(features_test))
````

````python
smape(target_test, pd.Series(target_test.median(), index=target_test.index))
````

````python
smape_rough = smape(target_test, grid_base.predict(features_test))
````
